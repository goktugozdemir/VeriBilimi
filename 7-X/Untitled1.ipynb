```python
data3 = (data.corr()>0.25) | (data.corr()<-0.25)
data.corr()[data3]
```

```python
data['1'] = (data['INDEPEND']*data["SEX"])
data['2'] = (data['LHIPPO']+data['RHIPPO'])
data['3'] = (25-data['NACCMMSE'])
data[['1','2','3','CDRGLOB']].corr()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score
from sklearn.naive_bayes import BernoulliNB
from sklearn.metrics import classification_report

data4=pd.concat([data['1'], data['2'],data['3']], axis=1)
br=BernoulliNB() 
X1, X2, Y1, Y2 = train_test_split(data4, data["CDRGLOB"], random_state=0, train_size=0.8,test_size=0.2)
br.fit(X1,Y1)                                                                
y3 = br.predict(X2)
param_grid = {'alpha': [1,5,10,15,20],                                       
              'binarize': [1,2,3],
              'fit_prior' : [True,False]}

grid = GridSearchCV(br, param_grid, cv=15)                             
grid.fit(X1, Y1)                                                                                              
best_grid = grid.best_estimator_                                      
grid_best = best_grid.fit(X1, Y1).predict(X2)                                 

accuracy_score(Y2,y3)
accuracy_score(Y2,grid_best)
classification_report(Y2, y3)
```
