```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def grad(x):
    return 2*x + 2

def gradient_descent(x, lr, max_iter):
    x_list = []
    for i in range(max_iter):
        grad_x = grad(x)
        x -= lr * grad_x
        x_list.append(x)
    return x_list

x = 10
max_iter = 1000

lr_list = [0.001, 0.01, 0.0001, 0.1]

iterations = np.arange(1, max_iter+1)

for lr in lr_list:
    x_list = gradient_descent(x, lr, max_iter)
    plt.plot(iterations, x_list)
    print(f"{lr}  {x_list[-1]}")


plt.legend(lr_list)
plt.show()
```
