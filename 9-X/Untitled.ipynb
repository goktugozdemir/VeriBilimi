```python
import pandas as pd
df = pd.read_excel("mri.xlsx")
df.head()
```

```python
df2=df.loc[:, 'NACCMMSE':'SEX']
dfn=df["NACCBRNV"]
dfnac=df["NACCAGEB"]
for i in range(len(dfn)):
  df2.iloc[i]=df2.iloc[i]/dfn.iloc[i]
  df2.iloc[i]=df2.iloc[i]*dfnac.iloc[i]
df3=df.loc[:, :'DROPACT']
df=pd.concat([df3, df2], axis=1, join='inner')
df
```

```python
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import classification_report
import numpy as np

y = df["CDRGLOB"]
X = df2
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=25)
gnb = GaussianNB()
y_pred = gnb.fit(X_train, y_train).predict(X_test)
accuracy_score(y_test, y_pred)

param_grid = {'var_smoothing': np.logspace(0,-9, num=100)}
grid = GridSearchCV(estimator=gnb, param_grid=param_grid, cv=5)
grid.fit(X_train, y_train);
grid.best_params_
gnb = grid.best_estimator_
my_predictions = gnb.fit(X_train,y_train).predict(X_test)
print(classification_report(y_test, my_predictions))
```
